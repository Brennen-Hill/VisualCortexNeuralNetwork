```python
import os
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

```python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
print(f"Using {device} device")
```

```python
class OurNeuralnetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super().__init__()
        # First layers transforms the input
        self.V1 = nn.Linear(input_dim, hidden_dim)
        self.ThickStripe = nn.Linear(hidden_dim, hidden_dim)

        # MT layer takes combined outputs of V1 and ThickStripe
        self.MT = nn.Linear(2 * hidden_dim, hidden_dim)

        self.VIP = nn.Linear(hidden_dim, hidden_dim)

        self.MST = nn.Linear(hidden_dim * 2, hidden_dim)

        self.Interstripe = nn.Linear(hidden_dim, hidden_dim)

        self.ThinStripe = nn.Linear(hidden_dim, hidden_dim)

        self.LIP = nn.Linear(hidden_dim, hidden_dim)

        self.V4 = nn.Linear(hidden_dim * 3, hidden_dim)

        self.PIT = nn.Linear(hidden_dim * 3, hidden_dim)

        self.CIT = nn.Linear(hidden_dim * 2, hidden_dim)

        self.SevenA = nn.Linear(hidden_dim * 2, hidden_dim)

        self.AIT = nn.Linear(hidden_dim * 2, output_dim)

    def forward(self, input):
        V1 = self.V1(input)

        ThickStripe = self.ThickStripe(V1)

        MT = self.MT(
            torch.cat((V1, ThickStripe), dim=1)
        )

        VIP = self.VIP(MT)

        MST = self.MST(
            torch.cat((MT, VIP), dim=1)
        )

        Interstripe = self.Interstripe(V1)

        ThinStripe = self.ThinStripe(V1)

        LIP = self.LIP(MST)

        V4 = self.V4(
            torch.cat((MT, Interstripe, ThinStripe), dim=1)
        )

        PIT = self.PIT(
            torch.cat((V4, MST, LIP), dim=1)
        )

        CIT = self.CIT(
            torch.cat((PIT, V4), dim=1)
        )

        SevenA = self.SevenA(
            torch.cat((MST, LIP), dim=1)
        )

        AIT = self.AIT(
            torch.cat((CIT, SevenA), dim=1)
        )
        return AIT
```
